```python


import pandas as pd
import re
import string

# Load the Blog Authorship Corpus dataset into a Pandas DataFrame
df=pd.read_csv('C:/Users/Admin/Downloads/Telegram Desktop/Book1.csv')
df

```

```python

# Define a function to clean the text data
def clean_text(text):
    # Remove punctuation
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
    # Convert to lowercase
    text = text.lower()
    # Remove numbers
    text = re.sub(r'\d+', '', text)
    # Remove extra whitespace
    text = re.sub(' +', ' ', text)
    return text

# Apply the clean_text function to the 'text' column of the DataFrame
df['clean_text'] = df['text'].apply(clean_text)
```

```python

def generate_ngrams(text, n):
    # Split the text into words
    words = text.split()
    # Generate n-grams
    ngrams = []
    for i in range(len(words)-n+1):
        # Check if the n-gram contains previously taken words in other n-grams
        if not set(words[i:i+n]).intersection(set([ngram.split()[j] for ngram in ngrams for j in range(len(ngram.split()))])):
            ngram = ' '.join(words[i:i+n])
            ngrams.append(ngram)
    return ngrams
```

```python


# Apply the generate_ngrams function to the 'clean_text' column of the DataFrame to generate and count ngrams
df['unigrams'] = df['clean_text'].apply(lambda x: generate_ngrams(x, 1))
df['bigrams'] = df['clean_text'].apply(lambda x: generate_ngrams(x, 2))
df['trigrams'] = df['clean_text'].apply(lambda x: generate_ngrams(x, 3))
```

```python
# Display the first few rows of the DataFrame with the new columns
print(df[['text', 'clean_text', 'unigrams', 'bigrams', 'trigrams']].head())

# Count the number of words in each n-gram type
df['unigram_word_count'] = df['unigrams'].apply(lambda x: len(x))
df['bigram_word_count'] = df['bigrams'].apply(lambda x: len(x))
df['trigram_word_count'] = df['trigrams'].apply(lambda x: len(x))

# Count the total number of words in the 'text' column
df['total_word_count'] = df['text'].apply(lambda x: len(x.split()))

# Display the first few rows of the DataFrame with the new columns
df
```
